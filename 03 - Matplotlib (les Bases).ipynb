### Graphiques simples

```python
X = np.linspace(0, 2, 10)
y = X**2

#plt.plot(X, y) # courbe simple
plt.plot(X, y) #, ls='dashed')  # dashed = '--'
plt.show()
```

```python
plt.scatter  # tab => compléter la fonction,  shift(maj) + tab => ouvre le détail de la fct
```

```python
plt.scatter(X, y) # nuage de points
plt.show()
```

### Styles Graphiques
Il existe beaucoup de styles a ajouter aux graphiques. Voici les plus importants a retenir :
- **c** : couleur de la ligne
- **lw** : epaisseur de la ligne (pour les graphiques plot)
- **ls** : style de la ligne (pour les graphiques plot) , possibilités = {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
- **size** : taille du point (pour les graphiques scatter)
- **marker** : style de points (pour les graphiques scatter)
- **alpha** : transparence du graphique



```python
plt.plot(X, y, c='red', lw=4, ls='--', alpha=0.8)
plt.show()
```

## 2. Cycle de vie d'une figure
Pour créer des figures proprements, on doit suivre le cycle de vie suivant :
1. **plt.figure**(figsize=())
2. **plt.plot()**
3. Extras (titre, axes, legendes)
4. **plt.show()** si IDE autre que Jupyter

```python
X = np.linspace(0, 2, 10)

plt.figure() # Création d'une figure - écrire ceci pour chaque figure à afficher
plt.plot(X, X**2, label='quadratique') # premiere courbe avec son label ( pour legend à la fin)
plt.plot(X, X**3, label='cubique') # deuxieme courbe avec son label
# Extra information
plt.title('figure 1') # titre
plt.xlabel('axe x') # label de l'axe x
plt.ylabel('axe y') # label de l'axe y
plt.legend() #   encadré legend() en dernier avec les définitions des labels dans les plot avant

plt.savefig('data/figure.png') # sauvegarde la figure dans le repertoire de travail
plt.show() # affiche la figure
```

## 3. Subplots
Les subplots sont un autre éléments a ajouter pour créer plusieurs graphiques sur une meme figure (max 9 figures)

```python
plt.figure(figsize=(12, 8))  # taille de la figure 12 inch en largeur et 8 inch en hauteur (x2,54cm)
x = np.linspace(-10, 10, 30)  # définir une ligne de 30 points x sur l'intervalle -10 ; +10
plt.subplot(2, 1, 1)        # 1er suplot définit
plt.plot(x, x**2, c='red')  # figure du 1er subplot
plt.subplot(2, 1, 2)       # 2eme subplot définit
plt.plot(x, x**3, c='blue')    # PS: le choix des valeurs des axes se fait automatiquement
```

## 4. Méthode orientée objet

```python
fig, ax = plt.subplots(2, 1, sharex=True) # partage le meme axe pour les subplots
x = np.linspace(-10, 10, 50)  # définir une ligne de 50 points x sur l'intervalle -10 ; +10
y = x**2
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```
