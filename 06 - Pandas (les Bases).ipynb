### PS : si on voulait utiliser matplotlib, ce serait plus joli et modifiable en graph mais plus complexe √† coder


```python
import matplotlib.pyplot as plt

# Cr√©er la figure et les axes
plt.figure(figsize=(10, 6))

# Tracer l'histogramme
plt.hist(data['age'].dropna(), bins=30, edgecolor='black')

# Ajouter un titre et des labels aux axes
plt.title('R√©partition des √¢ges', fontsize=14)
plt.xlabel('√Çge', fontsize=12)
plt.ylabel('Fr√©quence', fontsize=12)

# Afficher le graphique
plt.show()
```

## 3. Groupby() et value_counts()

```python
data.groupby(['sex']).mean()  # regrouper le dataframe par 'sex' Et faire la moyenne mean() - pclass=1, 2, 3 mean(pclass) peu utile
```

```python
data.groupby(['sex', 'pclass']).mean()  # ce regroupement est mieux ! moyenne par sex et par class
```
